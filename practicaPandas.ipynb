```python
data =  {
    'Ano': info.Ano,
    'PIB': info.PIB,
    'Pais': info.Pais, 
    'Uso_de_Internet': info["Uso de energia"]
}

data
```

```python
df = pd.DataFrame(data = data, columns=['Ano','PIB','Pais', 'Uso_de_internet']) 

```

```python
df[(df.PIB<5.479006e+10) & (df.Ano=="2000-12-01")]
```

```python
agrupado = df.groupby(['Pais'])['PIB'].transform('mean')
agrupado
df['Media PIB'] = agrupado
df.head()
```

```python
import matplotlib as mpl 
import matplotlib.pyplot as plt
```

```python
print(df)
```

```python
dfg = df.groupby(['Pais'])['PIB'].mean().reset_index()
dfg.columns = ['Pais', 'PIB']
dfg2 = dfg.sort_values('PIB',ascending=False).head(5)
dfg2.index = dfg2['Pais']
print(dfg2)


dfg2.plot.pie(
    y = 'PIB',
    legend=False,
    autopct='%1.1f',
    title="TOP 5 de los paises con PIB promedio mas alto"
)


plt.show()
```

```python
dfg = df.groupby(['Pais'])['PIB'].mean().reset_index()
dfg.columns = ['Pais', 'PIB']
dfg2 = dfg.sort_values('PIB',ascending=True).head(5)
dfg2.index = dfg2['Pais']
print(dfg2)


dfg2.plot.pie(
    y = 'PIB',
    legend=False,
    autopct='%1.1f',
    title="TOP 5 de los paises con PIB promedio mas bajo"
)


plt.show()
```
